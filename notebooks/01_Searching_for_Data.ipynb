```python
import razdel

from corus import load_wiki
```

```python
from tqdm import tqdm
import logging

logging.getLogger().setLevel(logging.ERROR)
```

```python
YO = 'ё'

wiki = load_wiki('../data/ruwiki-latest-pages-articles.xml.bz2')
count = 0

with open('../data/yo-ruwiki.txt', 'w') as file, tqdm(
    wiki, 
    desc='Adding `Ё` sentences from Wiki records',
    dynamic_ncols=True, 
    colour='green'
) as progress:
    for record in progress:
        sentences = razdel.sentenize(record.text)
        for sentence in sentences:
            if YO in sentence.text.lower():
                file.write(sentence.text + '\n')
                count += 1
        progress.set_postfix(found=count)
```
